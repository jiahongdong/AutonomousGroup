```python
import cv2
from darkflow.net.build import TFNet
import numpy as np
import time
```

```python
option = {
    'model': 'cfg/tiny-yolo-voc-5class.cfg',
    'load': 6900,
    'threshold': 0.15
}
'''
option = {
    'model': 'cfg/tiny-yolo-voc.cfg',
    'load': 'bin/tiny-yolo-voc.weights',
    'threshold': 0.1
}'''
'''option = {
    'model': 'cfg/tiny-yolo-voc-5class.cfg',
    'load': 6900,
    'threshold': 0.01
}'''

tfnet = TFNet(option)
```

```python
colors = [tuple(255 * np.random.rand(3)) for i in range(100)]
capture = cv2.VideoCapture('Hess.avi')
```

```python

'''frames = 0'''
while (capture.isOpened()):
    '''frames += 1'''
    startTime = time.time()
    ret, frame = capture.read()
    '''ret is true when vid playing, false when its not'''
    results = tfnet.return_predict(frame)
    print(results)
    label = []
    topleft = []
    bottomright = []
    
    if ret:
        for color, result in zip(colors, results):
            '''zip makes a list where each item is a tuple (one color and one result)'''
            topleft = (result['topleft']['x'], result['topleft']['y'])
            bottomright = (result['bottomright']['x'], result['bottomright']['y'])
            label = result['label']
            confidence = result['confidence']
            frame = cv2.rectangle(frame,topleft, bottomright, color, 3)
            frame = cv2.putText(frame, label, topleft, cv2.FONT_HERSHEY_COMPLEX, 1, (255,255,255), 2)
        cv2.imshow('frame',frame)
        print('FPS {:.01f}'.format(1/(time.time()-startTime)))
        if cv2.waitKey(1) & 0xFF == ord('q'):
            '''if we hit the q key, it will end'''
            break
            
    else:
        capture.release()
        cv2.destoryAllWindows()
        break
```
